```r
seriesFolder = 'data'
modelFolder  = 'model'
resultFolder = 'testResult'
```

```r
#select all model
modelsLake = list.files(path = modelFolder, pattern = NULL, all.files = FALSE, full.names = TRUE, recursive = FALSE)
```

```r
mda  = matrix(nrow=0, ncol=65)
mddl = matrix(nrow=0, ncol=65)
```

```r
fourier.models = modelsLake[grepl("*Fourier*", modelsLake)]
fourier.mda    = c()
fourier.mddl   = c()
for(m in fourier.models){
    load(m)
    fourier.mda  = c(fourier.mda,  model$mda)
    fourier.mddl = c(fourier.mddl, model$mddl)
}
mda  = rbind(mda,  fourier.mda)
mddl = rbind(mddl, fourier.mddl) 
```
